4. Describe how dictionaries store data.
  - Python dictionaries store data as key-value pairs within a hash table structure. This design enables efficient retrieval and manipulation of data based on unique keys.
    - Key aspects of dictionary data storage:
      - Key-Value Pairs: Each item in a dictionary consists of a key and its associated value. Keys must be unique and immutable data types (e.g., strings, numbers, tuples), while values can be of any data type and can be duplicated.
      - Hash Table Implementation: Dictionaries are implemented using hash tables. When a key-value pair is added, the key's hash value is calculated. This hash value determines the location (or "bucket") within the hash table where the key-value pair will be stored.
      - Hash Collisions: Multiple keys might produce the same hash value, leading to a "hash collision." Python's dictionary implementation handles these collisions, typically by using a technique like open addressing or chaining to store multiple key-value pairs within the same bucket or a linked structure.
      - Dynamic Resizing: As more key-value pairs are added, the dictionary's underlying hash table may need to be resized to maintain efficient performance and reduce the likelihood of excessive collisions. This involves creating a larger hash table and re-hashing existing elements.
      - Ordered Dictionaries (Python 3.7+): Since Python 3.7, dictionaries maintain insertion order. This means that the order in which items are added to the dictionary is preserved when iterating over its contents. In earlier versions, dictionaries were unordered.
    - This hash-based approach allows for average O(1) (constant time) complexity for operations like adding, retrieving, and deleting key-value pairs, making dictionaries a highly efficient data structure for many applications.

8. What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table is a data structure that stores key-value pairs by using a hash function to compute an index into an array, allowing for very fast data lookup. Python's built-in dictionary (dict) is the language's implementation of a hash table.
  - How Hash Tables Work
    - A hash table uses a special function called a hash function to convert a key into an integer, known as a hash code. This hash code directly corresponds to an index (or "bucket") in an underlying array where the value is stored. This process allows for average-case time complexity of O(1) for insertions, deletions, and lookups, meaning the operation time is constant and independent of the data size.
  - Dictionaries as Hash Tables
    - Python's dictionaries use this exact hash table mechanism to provide their efficient key-value functionality. When we add a key-value pair to a dictionary, Python:
      - 1. Hashes the key to generate a hash code. The key must be of a hashable type, meaning it's immutable (e.g., strings, numbers, tuples)
      - 2. Determines the index in the internal array based on this hash code
      - 3. Stores the key-value pair at that index.
  - Handling Collisions
    - It's possible for two different keys to produce the same hash code, an event known as a hash collision. Python dictionaries handle collisions using a method called open addressing. If a collision occurs, instead of placing the new entry in the same occupied slot, the system "probes" or searches for the next available empty slot according to a specific sequence and places the new entry there. This ensures that all key-value pairs are stored and can be retrieved, even if their keys initially hash to the same index.

14. How does the “in” keyword work differently for lists and dictionaries?
  - The "in" keyword checks for values in a list and keys in a dictionary. The primary difference lies in performance; dictionary lookups are significantly faster because they use a hash table, while lists require iterating through each element.
  - Lists: Linear Search for a Value
      - When using the in operator on a list, Python performs a linear search. It checks each element sequentially from the beginning until it finds a match or reaches the end of the list.
        - Time Complexity: O(n) This is known as O(n) time complexity, meaning the search time is directly proportional to the number of items (n) in the list. For large lists, this can be slow.
  - Dictionaries: Hashed Search for a Key
      - For dictionaries, the in operator checks for the presence of a key, not a value. This process is highly efficient because dictionaries are implemented as hash tables.
          - Time Complexity: O(1) On average, this gives it an O(1), or constant time, complexity. This means the search time remains roughly the same regardless of the dictionary's size, making it much faster for large datasets. In rare cases involving many hash collisions, the performance can degrade to O(n).

17.  Describe the time complexity of accessing elements in a dictionary
  - The time complexity of accessing an element in a dictionary is O(1) on average, but can be O(n) in the worst-case scenario. This efficiency is due to their underlying hash table implementation.
  - Average Case: O(1) (Constant Time)
  - Accessing an element in a dictionary is typically a very fast, constant-time operation. This is because dictionaries use a hash function to convert the key into a memory address. This allows for direct access to the value, regardless of the size of the dictionary. The process involves:
    - Hashing the key: A hash value (an integer) is computed from the key.
    - Direct lookup: This hash value is used as an index to find the element in memory.
  - Worst Case: O(n) (Linear Time)
  - In rare situations, the time complexity can degrade to O(n), where 'n' is the number of elements in the dictionary. This occurs because of hash collisions, which happen when two different keys produce the same hash value. When a collision occurs, the dictionary must store the colliding items in the same memory slot, often using a secondary data structure like a linked list. To find the correct element, it may have to search through all the items in that slot, which in the worst possible scenario could be all 'n' items in the dictionary.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?
  - Historically, Python dictionaries were unordered due to their hash table implementation, which prioritizes fast key-based lookups over insertion sequence. Since Python 3.7, this is no longer true; dictionaries now officially preserve insertion order.
    - Why Dictionaries Were Unordered
    - Python dictionaries are built on a data structure called a hash table or hash map. This system works by converting a key into a unique integer called a hash. This hash determines where the key-value pair is stored in memory. Because the storage location is based on the key's hash, not when it was added, older Python versions (before 3.6) could not guarantee any specific order when iterating through the dictionary.
    - Effect on Data Retrieval
      - The "unordered" nature was a direct consequence of optimizing for retrieval speed
      - Fast Lookups Instead of searching through items one by one (like in a list), the hash table allows Python to go directly to the memory location associated with the key's hash. This makes retrieving a value by its key extremely fast, with an average time complexity of O(1), meaning the lookup time remains constant regardless of the dictionary's size.
      - No Index Access Because the order was not guaranteed, you could not access elements by a numerical position or index, like my_dict[0]. Data retrieval is done exclusively using the key, for example, my_dict['name'].
    - The Modern Python Dictionary: Ordered by Default
    - Beginning with Python 3.6 (as an implementation detail) and officially guaranteed in Python 3.7 and later, standard dictionaries now remember the order in which items were inserted. This change was the result of a new, more memory-efficient dictionary implementation that also happened to preserve insertion order as a side effect.

    - A dictionary stores data as key-value pairs and is optimized for fast lookups when you know the key.
    - Access Method: Data is accessed using a unique, immutable key (like a string or number) to look up its associated value (e.g., my_dict['name'])
    - Performance: Dictionaries use hash tables internally, which makes retrieving a value by its key extremely efficient. The average time complexity for a lookup is O(1), regardless of the dictionary's size.

```python
# 18.  Write a code to remove the element 4 from the set {1, 2, 3, 4}.
# Ans.
my_set = {1, 2, 3, 4}
print(f"Original Set:{my_set}")
my_set.discard(4)
print(f"After Discarding '4' in set :{my_set}")
# The key difference is that remove() will raise a KeyError if the element to be removed is not found in the set, while discard() will do nothing in that case, without raising an error.
```

```python
# 19.  Write a code to find the union of two sets {1, 2, 3} and {3, 4, 5}.
# Ans.
# Define the two sets
set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Method 1: Using the | operator
union_set_operator = set1 | set2
print(f"Union using '|' operator: {union_set_operator}")

# Method 2: Using the .union() method
union_set_method = set1.union(set2)
print(f"Union using '.union()' method: {union_set_method}")
```

```python
# 20. Write a code to find the intersection of two sets {1, 2, 3} and {2, 3, 4}.
# Ans.
set1 = {1, 2, 3}
set2 = {2, 3, 4}

# Method 1: Using the & operator
intersection_set_1 = set1 & set2

# Method 2: Using the .intersection() method
intersection_set_2 = set1.intersection(set2)

print(f"The intersection of the sets using the & operator is: {intersection_set_1}")
print(f"The intersection of the sets using the .intersection() method is: {intersection_set_2}")
```

```python
# 26. Write a code to create a list, a tuple, and a dictionary, and print them all.
# Ans.
# 1. Create a list
# Lists are ordered, mutable (changeable), and allow duplicate elements.
# They are defined using square brackets [].
my_list = ["Apple", "Banana", "Cherry"]

# 2. Create a tuple
# Tuples are ordered, immutable (unchangeable), and allow duplicate elements.
# They are defined using parentheses ().
my_tuple = (2, 4, 6, 8, 10)

# 3. Create a dictionary
# Dictionaries are unordered (in Python versions before 3.7), mutable, and store data in key-value pairs.
# They are defined using curly braces {}.
my_dictionary = {
  "Brand": "Ford",
  "Model": "Mustang",
  "Year": 1964
}

# 4. Print each data structure
print("This is a list:")
print(my_list)
print("\nThis is a tuple:")
print(my_tuple)
print("\nThis is a dictionary:")
print(my_dictionary)
```

```python
# 29. Write a code to combine two dictionaries into one and print the result.
# Ans.
# Define two dictionaries
dict1 = {'a': 1, 'b': 2}
dict2 = {'b': 3, 'c': 4}
print(f"Dictionary1: {dict1}")
print(f"Dictionary2: {dict2}")
# Method 1: Using the update() method
# This modifies dict1 in place.
dict1_copy = dict1.copy() # Create a copy to avoid modifying the original
dict1_copy.update(dict2)
print(f"Using update(): {dict1_copy}")

# Method 2: Using dictionary unpacking (**)
# This creates a new dictionary.
merged_dict_unpacking = {**dict1, **dict2}
print(f"Using unpacking (**): {merged_dict_unpacking}")

# Method 3: Using the merge operator (|) (Python 3.9+)
# This also creates a new dictionary.
merged_dict_operator = dict1 | dict2
print(f"Using the | operator: {merged_dict_operator}")

```
